**Create a Contact book**

Contact Information: Store name, phone number, email, and address for each
contact.



```python
# Contact Book without GUI and Command-Line Output

# Initializing an empty list for storing the contacts
contacts = []



# Main loop for the contact book menu ... Easist part
while True:
    print("\n=== Danish's contact Book Menu ===")

    print("1. Add Contact")

    print("2. View Contacts")

    print("3. Search Contact")

    print("4. Update Contact")

    print("5. Delete Contact")

    print("6. Exit")



# Getting the  user  to input the choice in,  for menu choice...
    choice = input("Enter your choice (1-6): ")



# Adding the  Contact system....
    if choice == '1':

        name = input("Enter the contact's name: ")

        phone = input("Enter the contact's phone number: ")

        email = input("Enter the contact's email address: ")

        address = input("Enter the contact's address: ")



# Creating a dictionary for the contact and add it to the contacts list...
        contact = {"Name": name, "Phone": phone, "Email": email, "Address": address}
        contacts.append(contact)

        print("\nContact added successfully!")





# Creating the View Contacts system...
    elif choice == '2':

        if not contacts:
            print("No contacts available.")

        else:
            print("\n=== Contact List ===")


            for i, contact in enumerate(contacts, 1):
                print(f"{i}. Name: {contact['Name']}, Phone: {contact['Phone']}, Email: {contact['Email']}, Address: {contact['Address']}")




#Creating the Search Contact system ...
    elif choice == '3':
        search_name = input("Enter the name to search: ").lower()

        search_results = [contact for contact in contacts if search_name in contact['Name'].lower()]


        if search_results:
            print("\n=== Search Results ===")


            for i, contact in enumerate(search_results, 1):

                print(f"{i}. Name: {contact['Name']}, Phone: {contact['Phone']}, Email: {contact['Email']}, Address: {contact['Address']}")


        else:
            print("No matching contacts found.")




#Creating rhe Update Contact system...
    elif choice == '4':
        search_name = input("Enter the name to update: ").lower()

        update_results = [contact for contact in contacts if search_name in contact['Name'].lower()]



        if update_results:

            for contact in update_results:

                print(f"Updating contact: {contact['Name']}")

                contact['Name'] = input("Enter the new name: ")

                contact['Phone'] = input("Enter the new phone number: ")

                contact['Email'] = input("Enter the new email address: ")

                contact['Address'] = input("Enter the new address: ")

                print("Contact updated successfully!")
        else:
            print("No matching contacts found.")




# Creating the Delete Contact system...
    elif choice == '5':

        search_name = input("Enter the name to delete: ").lower()

        delete_results = [contact for contact in contacts if search_name in contact['Name'].lower()]



        if delete_results:
            for contact in delete_results:
                print(f"Deleting contact: {contact['Name']}")

                contacts.remove(contact)

                print("Contact deleted successfully!")


        else:
            print("No matching contacts found.")




    elif choice == '6':
        print("Progress Saved !!! ")
        break


    else:
        print("Invalid choice. Please enter a number between 1 and 6.")
```
