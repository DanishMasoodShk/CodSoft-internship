**A To-Do List application is a useful project that helps users manage
and organize their tasks efficiently. This project aims to create a
command-line or GUI-based application using Python, allowing
users to create, update, and track their to-do lists**

```python
#Following is the TO-DO list with command line.
#A seperate code for TO-DO list with functions and GUI invovled will me made in VS CODE(.Py file)

todo_list = []


while True:
    print("\n || ---\\ ---- **Welcome !! to .. Danish's TO-DO List !!** ---// ----|| ")

    print("\n|| -- Please select '1'to add a Task ")

    print("|| -- Please select '2'to View Tasks ")

    print("|| -- Please select '3'to Update Task ")

    print("|| -- Please select '4'to Delete Task ")

    print("|| -- Please select '0'to Exit the menu")

    choice = input("\n Please enter your choice (0-4): ")

    if choice == "0":
        print("\n OK ! Your progress has been saved.")
        break


    elif choice == "1":
        task_description = input("\n Please enter a task description: ")

        due_date = input("\n Please enter a due date (This Is Optional): ")

        task = {"description": task_description, "due_date": due_date}

        todo_list.append(task)

        print("\n Notice: Task added successfully!")


    elif choice == "2":

        if not todo_list:
            print("\n Notice: Sorry ! No tasks found.")

        else:
            for idx, task in enumerate(todo_list, start=1):

                print(f"{idx}. Description: {task['description']}")

                print(f"   Due Date: {task['due_date']}")

                print("-" * 30)

    elif choice == "3":
        if not todo_list:
            print("No tasks found.")


        else:
            for idx, task in enumerate(todo_list, start=1):

                print(f"{idx}. Description: {task['description']}")

                print(f"   Due Date: {task['due_date']}")

                print("-" * 30)

            task_index = int(input("\n Please enter the task number to update: ")) - 1

            if 0 <= task_index < len(todo_list):
                task_description = input("\n Please enter new task description (press Enter to keep the current one): ")

                task_due_date = input("\n Please enter new due date (press Enter to keep the current one): ")


                if task_description:
                    todo_list[task_index]['description'] = task_description


                if task_due_date:
                    todo_list[task_index]['due_date'] = task_due_date

                print("\n Notice : Task updated successfully!")


            else:

                print("\n Notice : Invalid task number. Please use valid entity. ")

    elif choice == "4":

        if not todo_list:
            print("\n Notice: Sorry ! No tasks found.")


        else:

            for idx, task in enumerate(todo_list, start=1):
                print(f"{idx}. Description: {task['description']}")

                print(f"   Due Date: {task['due_date']}")

                print("-" * 30)

            task_index = int(input("\n Please enter the task number to delete: ")) - 1
            if 0 <= task_index < len(todo_list):

                deleted_task = todo_list.pop(task_index)

                print(f"Task '{deleted_task['description']}' deleted successfully!")

            else:
                print("\n Notice: Sorry ! Invalid task number.Please use valid entity. ")

    else:
        print("\n Notice: Sorry ! Invalid choice. Please try again and use valid choice. ")
```
